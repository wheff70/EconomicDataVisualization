Importing data

```python
import pandas as pd
```

```python
econ_data = pd.read_csv("economicdata2021-2021.csv")
```

```python
econ_data.head()
```

```python
print(econ_data.columns)
```

```python
relevant_columns = ['Year', 'ISO Code', 'Countries', 'Government consumption',
                    'Government investment', 'Size of Government', 'Regulatory trade barriers',
                    'Financial openness', 'Controls of the movement of capital and people',
                    'Credit market regulations', 'Hiring regulations and minimum wage',
                    'Freedom to enter markets and compete', 'Regulation']
```

```python
clean_econ_data = econ_data.drop(econ_data.columns.difference(relevant_columns), axis=1)
clean_econ_data.head()
```

```python
clean_econ_data.to_csv('economic_data_cleaned.csv', index=False)
```
